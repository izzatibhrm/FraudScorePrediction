```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer

#Load dataset
df = pd.read_csv('clarity_underwritingv2.csv', low_memory=False)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
num_col = df.select_dtypes(include=[np.number,'float']).columns.to_list()
```

```python
num_col
```

```python
select = df[num_col]
```

```python
knn_imputer = KNNImputer(n_neighbors=5)
```

```python
imputed_data = knn_imputer.fit_transform(select)
```

```python
imputed_df = pd.DataFrame(imputed_data, columns=num_col)
```

```python
imputed_df
```

```python
imputed_df.isnull().sum()
```

```python
# prompt: export imputed_dfcombine with df.clarityFraudId into a csv file

import pandas as pd
df_combine = pd.concat([imputed_df, df['clarityFraudId']], axis=1)
df_combine.to_csv('clarity_underwritingv6.csv', index=False)
```
